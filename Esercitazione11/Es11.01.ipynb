```python
import os                              # import del modulo per interagire con il sistema operativo
# Forza lesecuzione su CPU (opzionale)
os.environ['CUDA_VISIBLE_DEVICES'] = ''  # imposta variabile dambiente per disabilitare GPU

import gc                              # garbage collector, per liberare memoria
import time                            # per pause temporali
import numpy as np                     # libreria per calcoli numerici
import matplotlib.pyplot as plt        # libreria per creazione grafici
import tensorflow as tf                # framework per deep learning
from tensorflow.keras import backend as K  # backend di Keras, per operazioni di sessione
from tensorflow.keras.models import Sequential  # modello sequenziale Keras
from tensorflow.keras.layers import Dense       # layer fully-connected

# Parametri della funzione target y = m x + b
m, b = 2, 1

def train_and_evaluate(Nepochs=30, Ntrain=500, sigma=0.0):
    """Allena un modello e restituisce storia, pesi, dati e predizioni."""
    # Ripulisce eventuale sessione Keras precedente
    K.clear_session()
    gc.collect()                   # forzata raccolta garbage per liberare memoria
    time.sleep(0.05)               # breve pausa per stabilit risorse

    # Genera i dati di training e validation
    np.random.seed(0)              # seme fisso per riproducibilit
    x_train = np.random.uniform(-1, 1, Ntrain)   # Ntrain punti uniformi in [-1,1]
    x_valid = np.random.uniform(-1, 1, 50)       # 50 punti di validazione
    x_valid.sort()                 # ordina x_valid per facilit di plotting
    # y_train  la retta m*x+b pi rumore gaussiano di dev. std. sigma
    y_train = np.random.normal(m * x_train + b, sigma)
    y_valid = np.random.normal(m * x_valid + b, sigma)

    # Stampa i parametri dellesperimento
    print(f"\n>>> TRAINING: epochs={Nepochs}, Ntrain={Ntrain}, noise={sigma}")

    # Definisce un modello sequential con un solo layer Dense(1)
    model = Sequential([Dense(1, input_shape=(1,))])
    # Compila il modello scegliendo lottimizzatore, la loss MSE e metriche MSE
    model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

    # Allena il modello sui dati di training, validando su quelli di validazione
    history = model.fit(
        x_train, y_train,            # dati di input e target
        epochs=Nepochs,              # numero di epoche
        batch_size=32,               # dimensione del mini-batch
        shuffle=True,                # mischia i dati ad ogni epoca
        validation_data=(x_valid, y_valid),  # dati di validazione
        verbose=1                    # stampa progresso ad ogni epoca
    )

    # Estrae i pesi della rete (w e b per un modello lineare)
    w_pred = model.get_weights()[0][0,0]  # peso del Dense
    b_pred = model.get_weights()[1][0]    # bias del Dense

    # Prepara punti equispaziati per predizione e calcola y_pred
    x_pred = np.linspace(-1, 1, 100)      # 100 punti in [-1,1]
    y_pred = model.predict(x_pred, verbose=0)  # predizioni

    # Libera risorse: distrugge il modello e pulisce sessione
    del model
    K.clear_session()
    gc.collect()
    time.sleep(0.05)

    # Restituisce tutti i risultati dellesperimento
    return history, w_pred, b_pred, x_train, y_train, x_valid, y_valid, x_pred, y_pred

# ----------------------------------------------------------
# Costruzione di una griglia 33 di subplot per i vari esperimenti
# ----------------------------------------------------------
fig, axes = plt.subplots(3, 3, figsize=(18, 15))  # 3 righe, 3 colonne
axes = axes.flatten()                              # appiattisce larray 2D in 1D

# Esperimento 1: variazione del numero di epoche
for i, epochs in enumerate([5, 30, 100]):
    # Allena e valuta per ogni valore di epochs
    hist, w, b_p, x_tr, y_tr, x_val, y_val, x_pr, y_pr = train_and_evaluate(
        Nepochs=epochs, Ntrain=500, sigma=0.0)
    ax = axes[i]                    # seleziona il subplot corrispondente
    ax.scatter(x_tr, y_tr, alpha=0.3, label='Train data')        # dati di training
    ax.plot(x_pr, y_pr, 'r-', label=f'Pred (w={w:.2f}, b={b_p:.2f})')  # retta predetta
    ax.plot(x_pr, m*x_pr + b, 'k--', label='True')               # retta target
    ax.set_title(f'Epochs = {epochs}')  # titolo del subplot
    ax.set_xlabel('x'); ax.set_ylabel('y')  # etichette assi
    ax.legend(); ax.grid(True)              # mostra legenda e griglia

# Esperimento 2: variazione del numero di punti di training
for i, n in enumerate([10, 100, 500]):
    hist, w, b_p, x_tr, y_tr, x_val, y_val, x_pr, y_pr = train_and_evaluate(
        Nepochs=30, Ntrain=n, sigma=0.0)
    ax = axes[3 + i]               # righe della seconda prova (subplot 3,4,5)
    ax.scatter(x_tr, y_tr, alpha=0.3, label='Train data')
    ax.plot(x_pr, y_pr, 'r-', label=f'Pred (w={w:.2f}, b={b_p:.2f})')
    ax.plot(x_pr, m*x_pr + b, 'k--', label='True')
    ax.set_title(f'Ntrain = {n}')
    ax.set_xlabel('x'); ax.set_ylabel('y')
    ax.legend(); ax.grid(True)

# Esperimento 3: variazione del livello di rumore
for i, noise in enumerate([0.0, 0.5, 1.0]):
    hist, w, b_p, x_tr, y_tr, x_val, y_val, x_pr, y_pr = train_and_evaluate(
        Nepochs=30, Ntrain=500, sigma=noise)
    ax = axes[6 + i]               # righe della terza prova (subplot 6,7,8)
    ax.scatter(x_tr, y_tr, alpha=0.3, label='Train data')
    ax.plot(x_pr, y_pr, 'r-', label=f'Pred (w={w:.2f}, b={b_p:.2f})')
    ax.plot(x_pr, m*x_pr + b, 'k--', label='True')
    ax.set_title(f'Noise  = {noise}')
    ax.set_xlabel('x'); ax.set_ylabel('y')
    ax.legend(); ax.grid(True)

# Adatta layout, salva e mostra la figura complessiva
plt.tight_layout()  
plt.savefig('linear_regression_exploration.png', dpi=300)  
plt.show()  
plt.close('all')  
```

## Analisi degli Effetti sui Parametri del Modello

Il codice esplora sistematicamente tre parametri chiave che influenzano la qualit dellapprendimento di una rete neurale semplice.

---

###  Numero di Epoche (iterazioni di addestramento)

**5 epoche:**
- `w = 0.37`, `b = 0.79` (target: w=2, b=1)
- Il modello non ha tempo sufficiente per apprendere i parametri corretti  **underfitting grave**
- La linea rossa di predizione  quasi orizzontale

**30 epoche:**
- `w = 1.87`, `b = 1.00`
- Predizione molto vicina al target
- Buon compromesso tra accuratezza e tempo

**100 epoche:**
- `w = 2.00`, `b = 1.00`
- Parametri perfetti. La linea di predizione coincide con quella del target
- Il modello converge completamente alla soluzione ottimale

 **Conclusioni:**
- Con poche epoche  il modello non apprende (underfitting)
- Aumentando le epoche  il modello converge
- Per problemi semplici, **2050 epoche** sono generalmente sufficienti

---

###  Numero di Punti di Addestramento

**10 punti:**
- `w = 0.52`, `b = 0.74`
- Dati insufficienti  **alta varianza**  predizione lontana dalla realt

**100 punti:**
- `w = 1.61`, `b = 0.89`
- Miglioramento evidente, ma la pendenza non  ancora corretta

**500 punti:**
- `w = 1.89`, `b = 1.00`
- Ricostruzione quasi perfetta. Pi dati  **meno incertezza**

 **Risultati alternativi coerenti:**
- 10 punti: `m = 1.80`, `b = 1.08`
- 100 punti: `m = 1.99`, `b = 1.02`
- 500 punti: `m = 2.00`, `b = 1.00`

 **Conclusioni:**
- Pochi dati causano predizioni imprecise
- Con pi dati, il modello **generalizza meglio**
- Per regressione lineare, **100+ punti** sono raccomandati

---

###  Effetto del Rumore ()

** = 0.0 (nessun rumore):**
- `w = 1.90`, `b = 1.00`
- I dati sono perfettamente allineati  ottimo fit

** = 0.5 (rumore moderato):**
- `w = 1.93`, `b = 0.94`
- Predizione ancora buona, ma lintercetta si discosta di pi

** = 1.0 (rumore elevato):**
- `w = 2.01`, `b = 0.89`
- Il modello "media" il rumore ma fatica con loffset

 **Risultati alternativi coerenti:**
-  = 0.0  `m = 2.00`, `b = 1.00`
-  = 0.5  `m = 2.02`, `b = 0.97`
-  = 1.0  `m = 1.96`, `b = 1.04`

 **Conclusioni:**
- Pi rumore  pi difficile stimare i parametri corretti
- Il modello approssima la **miglior linea media** attraverso i dati rumorosi
