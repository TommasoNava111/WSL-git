```python
import os                              # modulo per configurazione ambiente e variabili di sistema
# 0) Disabilita GPU e limita il multithreading per stabilit
os.environ['CUDA_VISIBLE_DEVICES'] = '' # disabilita l'uso di GPU
os.environ['OMP_NUM_THREADS'] = '1'      # limita il numero di thread OpenMP a 1
os.environ['MKL_NUM_THREADS'] = '1'      # limita il numero di thread MKL a 1
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'  # workaround per problemi su macOS/Anaconda

import gc                              # garbage collector per liberare memoria manualmente
import time                            # modulo per pause/timing
import numpy as np                     # libreria per array e operazioni numeriche
import matplotlib.pyplot as plt        # libreria per plotting
import tensorflow as tf                # framework di deep learning
from tensorflow.keras import backend as K  # backend di Keras per operazioni su sessione
from tensorflow.keras.models import Sequential  # modello sequenziale Keras
from tensorflow.keras.layers import Dense       # layer fully connected
from tensorflow.keras.optimizers import Adam    # ottimizzatore Adam

# Limita i thread interni di TensorFlow per riproducibilit e stabilit
tf.config.threading.set_intra_op_parallelism_threads(1)
tf.config.threading.set_inter_op_parallelism_threads(1)

# Seed per riproducibilit di NumPy e TensorFlow
np.random.seed(42)
tf.random.set_seed(42)

# Funzione polinomiale target: f(x) = 3x - 2x - 3x + 4
def true_function(x):
    return 3*x**3 - 2*x**2 - 3*x + 4

# Generazione dati
sigma = 0.2  # deviazione standard del rumore
x_train = np.random.uniform(-1, 1, 1000)  # 1000 punti di training uniformi in [-1,1]
y_train = np.random.normal(true_function(x_train), sigma)  # y_train = f(x_train) + rumore
x_valid = np.random.uniform(-1, 1, 200); x_valid.sort()    # 200 punti di validazione ordinati
y_valid = np.random.normal(true_function(x_valid), sigma)  # y_valid = f(x_valid) + rumore
x_test = np.linspace(-1.5, 1.5, 300)   # 300 punti di test equispaziati in [-1.5,1.5]
y_test = true_function(x_test)         # valori target esatti senza rumore

# Pulisci grafi Keras/TensorFlow precedenti
K.clear_session()
gc.collect()
time.sleep(0.05)

# Costruzione modello: 3 hidden layer con ReLU, infine output lineare
model = Sequential([
    Dense(64, activation='relu', input_shape=(1,)),  # primo layer: 64 neuroni, ReLU
    Dense(64, activation='relu'),                    # secondo layer: 64 neuroni, ReLU
    Dense(32, activation='relu'),                    # terzo layer: 32 neuroni, ReLU
    Dense(1)                                         # output layer: 1 neurone, lineare
])

# Compilazione modello: scelta di ottimizzatore, loss e metriche
model.compile(
    optimizer=Adam(learning_rate=0.001),  # optimizer Adam con learning rate 0.001
    loss='mse',                          # mean squared error come funzione di perdita
    metrics=['mae']                      # mean absolute error come metrica addizionale
)

print("Modello creato con successo. Inizio addestramento...")

# Addestramento del modello
history = model.fit(
    x_train, y_train,                 # dati di training
    epochs=200,                       # numero di epoche
    batch_size=64,                    # dimensione batch
    validation_data=(x_valid, y_valid),  # dati di validazione
    verbose=1                          # mostra il dettaglio delle epoche
)

print("Addestramento completato. Libero le risorse...")

# Valutazione del modello sui dati di test (anche fuori range)
test_loss, test_mae = model.evaluate(x_test, y_test, verbose=0)
print(f"\nTest MSE: {test_loss:.4f}, Test MAE: {test_mae:.4f}")

# Generazione delle predizioni
y_pred_train = model.predict(x_train, verbose=0)  # predizioni sui dati di training
y_pred_test  = model.predict(x_test, verbose=0)   # predizioni sui dati di test

# Rilascio risorse: distrugge il modello da memoria
del model
K.clear_session()
gc.collect()
time.sleep(0.05)

# Plot dei risultati
plt.figure(figsize=(15, 10))

# Subplot 1: Dati di training vs funzioni vera e predetta
plt.subplot(2, 2, 1)
plt.scatter(x_train, y_train, alpha=0.3, label='Training data')
plt.plot(x_test, y_test, 'k-', linewidth=2, label='True function')
plt.plot(x_test, y_pred_test, 'r-', linewidth=2, label='NN prediction')
plt.title('Confronto Predizioni e Funzione Reale')
plt.xlabel('x'); plt.ylabel('y')
plt.legend(); plt.grid(True)

# Subplot 2: Errore di predizione su punti di test
plt.subplot(2, 2, 2)
error = y_pred_test.flatten() - y_test
plt.plot(x_test, error, 'b-')
plt.fill_between(x_test, error, 0, alpha=0.1)
plt.title('Errore di Predizione (Predizione - Reale)')
plt.xlabel('x'); plt.ylabel('Errore')
plt.grid(True)

# Subplot 3: Andamento della loss durante laddestramento
plt.subplot(2, 2, 3)
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.title("Evoluzione della Loss durante l'Addestramento")
plt.xlabel('Epoch'); plt.ylabel('Loss (MSE)')
plt.legend(); plt.grid(True)

# Subplot 4: Zoom sui dati di training (regione x[-1,1])
plt.subplot(2, 2, 4)
plt.scatter(x_train, y_train, alpha=0.3, label='Training data')
plt.plot(x_test, y_test, 'k-', linewidth=2, label='True function')
plt.plot(x_test, y_pred_test, 'r-', linewidth=2, label='NN prediction')
plt.xlim([-1.1, 1.1])
plt.title('Zoom sulla Regione di Addestramento')
plt.xlabel('x'); plt.ylabel('y')
plt.legend(); plt.grid(True)

# Organizza i layout, salva la figura e la mostra
plt.tight_layout()
plt.savefig('risultati_polinomiale.png', dpi=300)
plt.show()
plt.close('all')

```

##  Architettura e impostazioni della rete neurale

- **Struttura del modello**:
  - 3 strati nascosti con **64, 64 e 32 neuroni**
  - **Funzione di attivazione ReLU** per gli strati nascosti
  - **Strato di output lineare** (senza attivazione): adatto per la regressione

- **Ottimizzazione**:
  - **Ottimizzatore**: Adam (robusto, adattivo)
  - **Learning rate**: 0.001
  - **Batch size**: 64 (buon compromesso tra efficienza e stabilit)

- **Addestramento**:
  - **200 epoche**: convergenza completa senza segni evidenti di overfitting
  - **Validation split**: 20% dei dati per il monitoraggio del modello
  - **Funzione di perdita (loss)**: MSE (Mean Squared Error), la pi usata nei problemi di regressione

---

##  Analisi dei risultati

1. **Confronto predizione e funzione reale**  
   Il modello  stato in grado di approssimare molto bene la funzione cubica \( f(x) = 4 - 3x - 2x^2 + 3x^3 \) nel dominio di training \([-1, 1]\), con uno scarto minimo tra valori reali e predetti.  
   I dati predetti seguono la curva reale in modo fluido, senza oscillazioni spurie.

2. **Errore di predizione (Predetto - Reale)**  
   Lerrore  vicino a 0 in tutto il dominio, segno che la rete ha appreso correttamente la struttura della funzione. I picchi agli estremi suggeriscono leggera difficolt ai bordi.

3. **Andamento della loss**  
   La curva della loss mostra una discesa rapida e stabile, con training e validation loss praticamente sovrapposte: assenza di overfitting, buon generalizzatore.

4. **Zoom sulla regione di training**  
   Nella finestra \([-1, 1]\) il modello segue perfettamente il trend cubico anche in presenza di rumore nei dati.

---

##  Relazione tra complessit, qualit del fit e generalizzazione

| Complessit (parametri) | Fit sui dati di training | Capacit di generalizzazione |
|-------------------------|--------------------------|------------------------------|
| Bassa                   | Underfitting: non apprende la forma cubica | Bassa, anche sui dati interni |
| Moderata (come qui)     | Ottimo fit della funzione reale | Buona generalizzazione sui dati test e leggera perdita fuori dominio |
| Alta                    | Overfitting: apprende anche il rumore | Scarsa generalizzazione, instabile fuori dominio |

**Conclusione**:  
- Aumentare la complessit (numero di parametri) migliora il fit sui dati di addestramento, ma pu portare a overfitting e cattiva generalizzazione se non si controlla con tecniche adeguate.  
- Un modello moderatamente complesso (come quello usato) riesce a **interpolare bene i dati** e mantiene una discreta **capacit predittiva anche su dati nuovi**, specialmente se questi restano nel dominio visto in fase di training.
