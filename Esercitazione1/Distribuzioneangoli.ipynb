```python
import matplotlib.pyplot as plt

# Inserisci il nome del file con i dati
nome_file = "angoli.txt"

# Legge i dati dal file
with open(nome_file, 'r') as file:
    dati = [float(linea.strip()) for linea in file if linea.strip()]

# Crea l'istogramma
plt.hist(dati, bins=100, edgecolor='black')

# Aggiunge etichette e titolo
plt.xlabel("Valori")
plt.ylabel("Frequenza")
plt.title("Istogramma dei dati")

# Mostra il grafico
plt.show()
```
