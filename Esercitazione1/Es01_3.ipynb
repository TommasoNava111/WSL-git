```python
import numpy as np
import matplotlib.pyplot as plt

# Caricamento dei dati dal file
# Il file contiene tre colonne:
# 1. Numero di lanci totali
# 2. Stima di π
# 3. Incertezza sulla stima
data = np.loadtxt("Es01_3.txt")

# Estrazione delle colonne
n_throws = data[:, 0]     # Numero cumulativo di lanci
pi_estimate = data[:, 1]  # Stime di π
uncertainty = data[:, 2]  # Incertezze statistiche (deviazione standard della media)

# Creazione del grafico con barre d'errore
plt.figure(figsize=(10, 6))
plt.errorbar(
    n_throws,          # Asse x: numero di lanci
    pi_estimate,       # Asse y: stime di π
    yerr=uncertainty,  # Barre d'errore verticali (incertezza)
    fmt='o-',          # Formato: punti con linee continue
    markersize=4,      # Dimensione dei marcatori
    capsize=3,         # Dimensione delle estremità delle barre d'errore
    color='royalblue', # Colore della linea
    ecolor='crimson',  # Colore delle barre d'errore
    label=r'Stima di $\pi$'  # Etichetta per la legenda (raw string)
)

# Linea orizzontale con il valore vero di π
plt.axhline(
    y=np.pi,
    color='green',
    linestyle='--',
    label=r'Valore vero di $\pi$'  # Raw string per evitare il warning su \p
)

# Etichette degli assi e titolo
plt.xlabel("Numero di lanci", fontsize=12)
plt.ylabel(r"Stima di $\pi$", fontsize=12)
plt.title(r"Stima di $\pi$ con il metodo dell'ago di Buffon", fontsize=14)

# Griglia e legenda
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=11)

# Migliora il layout e mostra il plot
plt.tight_layout()
plt.show()
```

## Analisi del grafico della stima di $ \pi $ con il metodo di Buffon

Il grafico mostra l’evoluzione della stima di $ \pi $ e della sua incertezza in funzione del numero di lanci nell’esperimento dell’ago di Buffon. Ecco le principali osservazioni:

### 🔍 Andamento della stima
- **Convergenza al valore vero**:  
  Si osserva chiaramente come la stima (linea blu) si avvicini progressivamente al valore vero di $ \pi $ (linea verde tratteggiata) all’aumentare del numero di lanci.  
  
- **Comportamento iniziale**:  
  Nelle prime fasi (fino a circa 10 000 lanci) la stima mostra forti oscillazioni, tipico comportamento di processi stocastici con piccoli campioni.

- **Stabilizzazione**:  
  Oltre i 50 000 lanci la stima si stabilizza attorno al valore vero, con fluttuazioni via via minori.

### 📊 Interpretazione delle barre d’errore
- **Riduzione dell’incertezza**:  
  Le barre d’errore (rosse) rappresentano l’incertezza statistica (deviazione standard della media). Si nota come la loro dimensione diminuisca con l’aumentare dei lanci, seguendo l’andamento atteso $ 1 / \sqrt{N} $.

- **Copertura statistica**:  
  Nella maggior parte dei casi, il valore vero di $ \pi $ rimane all’interno dell’intervallo di incertezza, dimostrando la validità della stima statistica.

### ⚖️ Verifica della formula di Buffon
Il risultato sperimentale conferma la relazione teorica:  
$$
P_{\text{intersezione}} \;=\; \frac{2\,L}{\pi\,d}
\quad\Longrightarrow\quad
\pi \;=\; \frac{2\,L}{P\,d}
$$
dove:
- $ L $ = lunghezza dell’ago  
- $ d $ = distanza tra le linee  
- $ P = \dfrac{N_{\text{hit}}}{N_{\text{thr}}} $ = frequenza relativa di intersezioni

### 💡 Conclusioni
1. **Legge dei grandi numeri**:  
   L’esperimento dimostra empiricamente come, all’aumentare del numero di prove, la stima converga al valore atteso teorico.  
2. **Importanza dell’incertezza**:  
   La corretta quantificazione dell’incertezza è cruciale per interpretare risultati sperimentali, specialmente con pochi dati.
